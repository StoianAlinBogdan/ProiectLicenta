```python
from qiskit import Aer, QuantumCircuit
import matplotlib.pyplot as plt
import math
import pandas as pd
%matplotlib inline
```

```python
qc = QuantumCircuit(5)
qc.ry( 0.022 * math.pi, 0)
qc.ry( 0.136 * math.pi, 1)
qc.ry( 0.682 * math.pi, 2)
qc.ry( 0.136 * math.pi, 3)
qc.ry( 0.022 * math.pi, 4)
qc.measure_all()
qc.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc, shots=1000000).result()
counts = result.get_counts()
#counts2 = [(x / max(counts.values())) for x in counts.values()]
counts2 = {x : y / max(counts.values()) for x in counts.keys() for y in counts.values()}
plt.bar([int(x, 2) for x in counts2.keys()], counts2.values())
plt.show()
counts2
```
